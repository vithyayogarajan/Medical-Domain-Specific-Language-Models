```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("tail_wins_losses - 158_tail.csv")
```

```python
df
```

```python
df["Models"] = pd.Categorical(df["Models"],df.Models.unique())
```

```python
df_matrix = df.pivot("Models","Results","Score")
```

```python
fig = plt.figure(figsize=(9,9))
sns.set(font_scale=2.8)
r = sns.heatmap(df_matrix, cmap='Greens' ,xticklabels=["wins", "draws", "losses"],yticklabels=False,annot=True, annot_kws={"size": 24},linewidths=2, linecolor='white')
r.set_title("Tail-end Labels, Level 2 ICD-9",fontsize=30)
r.set_ylabel('')    
r.set_xlabel('')
```
